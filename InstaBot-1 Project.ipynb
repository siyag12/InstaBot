### Like/Unlike 30 posts of dilsefoodie

```python
search=driver.find_element(By.CLASS_NAME,"_aaw8")
search.click()
searchbox=driver.find_element(By.CLASS_NAME,"_aawh")
searchbox.send_keys("dilsefoodie")
```

```python
base_url="https://www.instagram.com/"
search_results=driver.find_element(By.XPATH,"//div[contains(@class,'_abnx ')]//a[contains(@class,'oajrlxb2')]")
data=bs(search_results.get_attribute("outerHTML"),"html.parser")
url=base_url+data.a["href"]
driver.get(url)
```

```python
i=0
while i<30:
    posts=driver.find_elements(By.XPATH,"//div[contains(@class,'_aa-i')]//div[contains(@class,'_aabd')]")
#     print(i.get_attribute("outerHTML"))
    data=bs(posts[i].get_attribute("outerHTML"),"html.parser")
    url=base_url+ data.a["href"]
    driver.get(url)
    time.sleep(3)
    like_button=driver.find_element(By.XPATH,"//section[contains(@class,'_aamu')]//button[contains(@class,'_abl-')]")
    like_diff=driver.find_element(By.XPATH,"//section[contains(@class,'_aamu')]/span/button/div")
    a=like_diff.get_attribute("outerHTML")
    data1=bs(a,"html.parser")
    if data1.svg["aria-label"]=="Like":
        like_button.click()
    else:
        print("Already Liked")
    time.sleep(2)
    driver.back()
    i+=1
    time.sleep(3)
```

```python
i=0
while i<30:
    posts=driver.find_elements(By.XPATH,"//div[contains(@class,'_aa-i')]//div[contains(@class,'_aabd')]")
#     print(i.get_attribute("outerHTML"))
    data=bs(posts[i].get_attribute("outerHTML"),"html.parser")
    url=base_url+ data.a["href"]
    driver.get(url)
    time.sleep(3)
    like_button=driver.find_element(By.XPATH,"//section[contains(@class,'_aamu')]//button[contains(@class,'_abl-')]")
    like_diff=driver.find_element(By.XPATH,"//section[contains(@class,'_aamu')]/span/button/div")
    a=like_diff.get_attribute("outerHTML")
    data1=bs(a,"html.parser")
    if data1.svg["aria-label"]=="Unlike":
        like_button.click()
    else:
        print("Already Unliked")
    time.sleep(2)
    driver.back()
    i+=1
    time.sleep(3)
    break
```
